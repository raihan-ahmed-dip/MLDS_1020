```python
import requests
```

```python
url = 'https://api.exchangerate.host/timeseries?start_date=2020-01-01&end_date=2023-01-04&base=USD&symbols=BDT'
```

```python
url = 'https://api.exchangerate.host/convert?from=USD&to=BDT'
```

```python
response = requests.get(url)
data = response.json()
```

```python
print(data)
```

```python
print(data['rates'])
```

```python
bdt_usd_rates = data['rates']
for key, val in bdt_usd_rates.items():
    print(key, ' - ', val['BDT'])
```
