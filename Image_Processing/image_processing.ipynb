# Image Processing

## 1. Imports

```python
from skimage import io
from skimage import color
from skimage import transform
from skimage import exposure
from skimage import filters
from skimage import feature
```

```python
import os
import matplotlib.pyplot as plt
```

## 3. Colors Transformation

```python
image = io.imread('data//bird.jpg', as_gray=True)
image
```

```python
io.imshow(image)
```

```python
image = io.imread('data//bird.jpg')
io.imshow(image)
```

```python
# Color Conversion
hsv_img = color.rgb2hsv(image)
hsv_img
```

```python
io.imshow(hsv_img, cmap='hsv')
```

```python
rgb_img = color.hsv2rgb(hsv_img)
io.imshow(rgb_img) 
```

```python
plt.imshow(hsv_img[:, :, 0], cmap='hsv')
```

## 4. Shape Transformation

```python
image.shape
```

```python
resized_img = transform.resize(image, output_shape=(128, 128))
resized_img.shape
```

```python
io.imshow(resized_img)
```

```python
rescaled_img = transform.rescale(image, 1/3)
resized_img.shape
```

```python
io.imshow(rescaled_img)
```

```python
rotate_img = transform.rotate(image, 90, resize=True)
io.imshow(rotate_img)
```

## 5. Exposure transformation

```python
sig_img = exposure.adjust_sigmoid(image, 0.3)
io.imshow(sig_img)
```

```python
log_image = exposure.adjust_log(image, 0.3)
io.imshow(log_image)
```

```python
gam_image = exposure.adjust_gamma(image, 0.4)
io.imshow(gam_image)
```

## 6. Filters

```python
blur_img = filters.gaussian(image, 3, multichannel=True)
io.imshow(blur_img)
```

## 7. Feature transformation

```python
vector = feature.hog(image, multichannel=True)
vector.shape
```

```python
vactor, img = feature.hog(image, multichannel=True, visualize=True)
```
